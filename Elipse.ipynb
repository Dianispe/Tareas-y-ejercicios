Estimar pi con elipse



```python
import random
import numpy as np
import matplotlib.pyplot as plt

def pi_mc(n):
    pi = 0
    lx = []
    ly = []
    counter = 0
    for i in range(n):
        x = random.random()
        y = random.random()
        a = random.random()
        b = random.random()
        if ((x)**2)/b**2 + ((y)**2)/a**2 <1:
            counter = counter + 1
            lx.append(counter)
            ly.append(4.0*float(counter)/(i + 1))
        
        pi = 4.0*float(counter)/(i + 1)
        
        e = np.absolute((pi - np.pi)/np.pi)
    print("pi =", pi)
    print("error =", e)
    
    fig = plt.figure(1, figsize=(8, 8))        
    plt.xlabel("Iteraciones")
    plt.ylabel("Estimación de $\pi$")
    plt.plot(lx, ly, color="blue", markersize=1)
    plt.show()

pi_mc(1000)
```

```python
# ESTE EJERCICIO NO SE LOGRÓ ):

import random
import numpy as np
import matplotlib.pyplot as plt

N = 1000
Aciertos = 0

XElipse=[]  
YElipse=[]  
XSquare=[]  
YSquare=[]  

for p in range(N):
      x = random.random()
      y = random.random()
      a = random.random()
      b = random.random()
      if (x**2/b**2 + y**2/a**2) <=1:
        Aciertos+=1
        XElipse.append(x)  
        YElipse.append(y)        
      else:
        XSquare.append(x)  
        YSquare.append(y)

Pi = 4*Aciertos/N

print("N=%d Aciertos=%d Pi=%.2f" %(N,Aciertos,Pi))

XLin=np.linspace(0,1)
YLin=[]
for x in XLin:
    YLin.append(np.sqrt(1-x**2))

fig = plt.figure(1, figsize=(8,8)) 
plt.axis("equal")                             
plt.grid(which="major")                        
plt.plot(XLin , YLin, color="red" , linewidth="1") 
plt.scatter(XElipse, YElipse, color="blue", marker   =".") 
plt.scatter(XSquare, YSquare, color="green"  , marker   =".") 
plt.title("Estimación de $\pi$ por el método de Monte Carlo")
plt.xlabel("X")
plt.ylabel("Y")
 
plt.show()
```
