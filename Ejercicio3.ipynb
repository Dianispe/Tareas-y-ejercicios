Tarea 1.
Estimar el valor de pi usando la función random

```python
from random import random
from statistics import mean,variance
from math import sqrt,pi
import numpy as np

nr=100000
l1=[ ]
l2=[ ]
for i in range(nr):
     x=random()
     y=random()
     if x**2+y**2<1:
          I1=1
     else:
          I1=0
     l1.append(I1)
     l2.append(x*y)
print("La media es: ",mean(l1), "y multiplicada por 4:", mean(l1)*4)
print("\nEl valor exacto de pi es: ",pi)
```

```python
import random
import matplotlib.pyplot as plt
import numpy as np
from random import random
from statistics import mean,variance
from math import sqrt,pi
import numpy as np

x = np.linspace(0, 1, 100)
y = np.sqrt(1 - x**2)                
l1=[]
l2=[]
for i in range(1000):
     x1=random()
     y1=random()
     l1.append(x1)
     l2.append(y1)


fig = plt.figure(1, figsize=(8, 8))        
plt.plot(x, y, color="pink", markersize=1) 
plt.scatter(l1, l2, s=20, c="yellow")
plt.title("Estimación de $\pi$")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
