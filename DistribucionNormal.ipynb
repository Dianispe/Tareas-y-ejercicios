```python
#Doce uniforme

import numpy as np
import matplotlib.pyplot as plt
import random
from scipy.stats import norm

def docun():
  r=0
  for i in range(12):
    r=r+random.random()
  r=r-6
  return r 

md=[]
n=100000

for i in range(n):
  md.append(docun())

plt.hist(md, range=(-10,10), bins=1000)
plt.title("Doce uniforme")
plt.show()
```

```python
# Gauss

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from random import random,randint

mu=randint(100,1000)
var=randint(1,10)
x1=mu+var
x2=mu-var

z1=(x1-mu)/var
z2=(x2-mu)/var

x=np.arange(z1,z2,0.0001)
x2=np.arange(-10,10,0.0001)
y=norm.pdf(x,0,1)
y2=norm.pdf(x2,0,1)
fig,ax=plt.subplots(figsize=(5,4))
plt.style.use("fivethirtyeight")
ax.plot(x2,y2)
ax.set_xlim([-5,5])
ax.set_title("Gauss")
plt.show()
```
